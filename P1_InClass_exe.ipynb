#### 2.3 布尔索引 - NumPy的“大杀器” 🔥 (重点 + 难点)

这是我们筛选数据的最强武器！一定要掌握。

**回顾：三步魔法**
1.  **提出条件** -> 得到**布尔地图** (`True`/`False`数组)
2.  **按图索骥** -> 把布尔地图放进索引 `[]`
3.  **得到结果** -> 只取出对应`True`的元素

````python
# 示例数据：一组应力测量值 (MPa)
stress_values = np.array([150, 210, 180, 250, 190, 220])
print(f"原始应力数据: {stress_values}")

# 1. 提出条件：哪些应力值超过了危险阈值 200 MPa？
is_dangerous_mask = stress_values > 200
print(f"是否超过200？ (布尔地图): {is_dangerous_mask}")

# 2. 按图索骥：筛选出所有危险的应力值
dangerous_stresses = stress_values[is_dangerous_mask]
print(f"筛选出的危险应力值: {dangerous_stresses}")
````

> #### 💡 布尔索引的应用非常广泛！

**组合条件 (难点！)**

我们需要找出那些**既不太高也不太低**的“正常”应力值（比如在160到210 MPa之间）。

````python
stress_values = np.array([150, 210, 180, 250, 190, 220])

# 条件1：应力 >= 160
condition1 = (stress_values >= 160)
# 条件2：应力 <= 210
condition2 = (stress_values <= 210)

# 组合条件：必须同时满足条件1 和 条件2
# ！！！注意：必须用 & 连接，并且每个条件都要用括号括起来 ！！！
normal_range_mask = condition1 & condition2

print(f"原始应力: {stress_values}")
print(f"是否 >= 160？: {condition1}")
print(f"是否 <= 210？: {condition2}")
print(f"是否在正常范围？: {normal_range_mask}")

normal_stresses = stress_values[normal_range_mask]
print(f"\n筛选出的正常应力值: {normal_stresses}")
````

#### 3.2 案例实战准备：线性方程组求解器 `np.linalg.solve()`

在我们进入最终的弹簧系统案例之前，还需要最后一块拼图：如何让NumPy帮我们解方程组 `Kx = F`？

NumPy的线性代数模块 `linalg` (Linear Algebra) 提供了一个非常方便的函数：`solve()`。

````python
# 假设我们有一个简单的方程组：
# 2x + 3y = 8
# 1x - 1y = 1

# 对应的矩阵形式 Ax = b 是：
# [[2, 3], [1, -1]] * [[x], [y]] = [[8], [1]]

# 1. 定义系数矩阵 A
A = np.array([
    [2, 3],
    [1, -1]
])

# 2. 定义结果向量 b
b = np.array([8, 1])

# 3. 调用求解器！
x_solution = np.linalg.solve(A, b)

print(f"系数矩阵 A:\n{A}")
print(f"结果向量 b: {b}")
print(f"\n求解得到的解向量 x (包含x和y的值): {x_solution}")

# 验证一下： A @ x_solution 是否等于 b ？
print(f"验证 A @ x: {A @ x_solution}")
print(f"是否接近 b? {np.allclose(A @ x_solution, b)}")
````

> **💡 关键：** `np.linalg.solve(系数矩阵, 结果向量)` 直接返回解向量。
> 
> 我们马上就要用这个函数来求解弹簧系统的位移了！

### 📍 **单元四：终极实战 - 弹簧系统分析 & AI调试 (约70分钟)**

#### 4.1 问题回顾

**系统**：墙-k1-m1-k2-m2-F
**已知**：k1=100, k2=120, F=50
**求解**：平衡位移 x1, x2

**我们已经推导出矩阵形式**：
$$ 
\begin{bmatrix} 
k_1+k_2 & -k_2 \\ 
-k_2 & k_2 
\end{bmatrix} 
\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} 
= 
\begin{bmatrix} 0 \\ F \end{bmatrix} 
$$

#### 4.2 编程求解 (一起动手！)

请大家跟着我一起，在下面的代码单元格里，一步步把求解过程写出来。

````python
import numpy as np
import matplotlib.pyplot as plt

# ====> 步骤 1: 定义已知参数 <====
k1 = 100.0
k2 = 120.0
F_ext = 50.0
print("--- 参数定义完成 ---")

# ====> 步骤 2: 构建刚度矩阵 K <====
# TODO: 根据上面的公式，填充这个2x2矩阵
K = np.array([
    [k1 + k2, -k2   ], 
    [-k2   ,  k2    ]  
])
print("--- 刚度矩阵 K ---")
print(K)

# ====> 步骤 3: 构建外力向量 F <====
# TODO: F作用在m2上，对应第二个自由度
F_vec = np.array([ 0 , F_ext ])
print("\n--- 外力向量 F ---")
print(F_vec)

# ====> 步骤 4: 求解位移向量 x <====
# TODO: 调用 np.linalg.solve()
x = np.linalg.solve(K, F_vec)
print("\n--- 求解完成 --- ")

# ====> 步骤 5: 输出结果 <====
print(f"x1 = {x[0]:.4f} m ({x[0]*1000:.2f} mm)")
print(f"x2 = {x[1]:.4f} m ({x[1]*1000:.2f} mm)")

# ====> 步骤 6: 数值验证 (计算残差) <====
# TODO: 计算 K @ x - F_vec
residual = K @ x - F_vec
print(f"\n--- 数值验证 (残差) ---")
print(f"残差向量: {residual}")
if np.allclose(K @ x, F_vec):
    print("✓ 残差接近零，计算精确！")
else:
    print("✗ 残差较大，请检查！")

# ====> 步骤 7: 物理验证 (检查m2力平衡) <====
# TODO: 计算弹簧2的力，看是否等于 F_ext
force_k2 = k2 * (x[1] - x[0])
print(f"\n--- 物理验证 (m2平衡) ---")
print(f"计算出的弹簧2弹力: {force_k2:.2f} N")
print(f"施加的外力: {F_ext:.2f} N")
if np.isclose(force_k2, F_ext):
    print("✓ 力平衡，物理验证通过！")
else:
    print("✗ 力不平衡，请检查模型！")

# ====> 步骤 8: 可视化结果 <====
print("\n--- 结果可视化 ---")
fig, ax = plt.subplots(figsize=(6, 5))
bars = ax.bar(['m1', 'm2'], x * 1000, color=['skyblue', 'salmon'])
ax.set_ylabel('位移 (mm)')
ax.set_title('弹簧系统平衡位移')
ax.grid(axis='y', linestyle=':')
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2.0, yval + 5, f'{yval:.1f}', 
            va='bottom', ha='center')
plt.show()
````

#### 4.3 AI 辅助调试实战 (处理潜在错误)

我们刚才求解 `Kx = F` 很顺利，但如果 `K` 矩阵有问题（比如行列式为0，称为**奇异矩阵**），`np.linalg.solve()` 就会报错。

**模拟一个错误场景：** 假设我们不小心把 `k1` 设为 `-k2` (这在物理上可能对应一个不稳定系统)。

````python
k1_bad = -120.0 # 假设k1设错了
k2_good = 120.0
F_vec = np.array([0, 50.0])

# 构建一个有问题的刚度矩阵
K_bad = np.array([
    [k1_bad + k2_good, -k2_good],
    [-k2_good,          k2_good]
])

print("有问题的刚度矩阵 K_bad:")
print(K_bad)

# 尝试求解 (这会报错！)
try:
    print("\n尝试求解...")
    x_bad = np.linalg.solve(K_bad, F_vec)
    print("求解结果: ", x_bad)
except np.linalg.LinAlgError as e:
    # except 块会在 try 块中发生错误时执行
    print("\n🔴 求解失败！")
    print(f"错误信息: {e}")
````
